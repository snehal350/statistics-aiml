```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Normal Distribution - bell curve (e.g. heights, marks, errors)
x = np.linspace(-4, 4, 100)
y = stats.norm.pdf(x, 0, 1)   # mean=0, std=1

plt.plot(x, y, color='blue')
plt.title("Normal Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.show()

# Binomial Distribution - number of successes in fixed trials (e.g. coin flips)
n, p = 10, 0.5
x2 = np.arange(0, n+1)
y2 = stats.binom.pmf(x2, n, p)

plt.bar(x2, y2, color='green')
plt.title("Binomial Distribution (10 coin flips)")
plt.xlabel("Number of Heads")
plt.ylabel("Probability")
plt.show()

# Example: probability of exactly 5 heads in 10 flips
print("P(5 heads in 10 flips) =", stats.binom.pmf(5, 10, 0.5))
```

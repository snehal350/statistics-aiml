```python
def bayes(prior, likelihood, false_alarm):
    """
    Return the posterior P(cause | evidence)

    prior: P(cause)
    likelihood: P(evidence | cause)
    false_alarm: P(evidence | not cause)
    """
    evidence = likelihood * prior + false_alarm * (1 - prior)
    return (likelihood * prior) / evidence


# Rare disease test from Worked Problem 1
post = bayes(prior=0.01, likelihood=0.99, false_alarm=0.05)

print(f"P(disease | positive) = {post:.4f}")
```

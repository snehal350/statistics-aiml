```python
import numpy as np
import matplotlib.pyplot as plt

# Bayes' Theorem: P(A|B) = [P(B|A) * P(A)] / P(B)

P_disease = 0.01                     # prior: 1% have disease
P_pos_given_disease = 0.95           # test sensitivity
P_pos_given_no_disease = 0.05        # false positive rate

P_positive = (P_pos_given_disease * P_disease) + \
             (P_pos_given_no_disease * (1 - P_disease))

P_disease_given_positive = (P_pos_given_disease * P_disease) / P_positive

print(f"P(Disease | Positive Test) = {P_disease_given_positive:.4f}")

# Quick plot: posterior vs prior
priors = np.linspace(0.001, 0.5, 100)
posteriors = (P_pos_given_disease * priors) / \
             (P_pos_given_disease * priors + P_pos_given_no_disease * (1 - priors))

plt.plot(priors, posteriors, color='darkred')
plt.xlabel('Prior P(Disease)')
plt.ylabel('Posterior P(Disease | Positive)')
plt.title("Bayesian Update")
plt.grid(True)
plt.show()
```

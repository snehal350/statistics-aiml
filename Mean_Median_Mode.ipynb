```python
import numpy as np
from scipy import stats

data = [25, 35, 35, 40, 42, 45, 45, 45, 50, 52, 55, 66]

# Arithmetic Mean
Mean = np.mean(data)
print(f"Mean: {Mean:.2f}")

# Median
Median = np.median(data)
print(f"Median: {Median:.2f}")

# Mode
mode_r = stats.mode(data, keepdims=True)
print(f"Mode: {mode_r.mode[0]} (freq={mode_r.count[0]})")

# Weighted Mean
weights = [1, 1, 2, 3, 2, 4, 4, 4, 2, 1, 1, 1]
W_mean = np.average(data, weights=weights)
print(f"W.Mean: {W_mean:.2f}")
```

Libraries:
1. Numpy- Fast array operations
2. Scipy.stats- Mode funtion
3. numpy.average - weighted mean
